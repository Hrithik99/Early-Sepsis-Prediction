```python
def load_data_from_pickle(file_path):
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    return data
```

```python
def read_data(filepath, df):
    """
    Reusable function to read PSV data from source directory.
    
    Parameters:
        filepath (str): The path to the directory containing PSV files.
        df (pd.DataFrame): The DataFrame to append the read data to.
        
    Returns:
        pd.DataFrame: The DataFrame containing all the read data.
    """
    count = 0
    rows = 0
    data_frames = []
    
    for filename in os.listdir(filepath):
        if filename.endswith(".psv"):
            with open(filepath + filename) as openfile:
                patient = filename.split("p")[1]
                patient = patient.split(".")[0]

                file = pd.read_csv(openfile, sep="|")
                file['Patient_ID'] = patient
                
                file = file.reset_index()
                file = file.rename(columns={"index": "Hour"})
                print(file)
                
                data_frames.append(file)
                rows += file.shape[0]
                count += 1
                
        # Print progress after 10k files
        if count % 10000 == 0:
            print("Progress || Files: {} || Number of items: {}".format(count, rows))
    
    # Concatenate all collected DataFrames
    df = pd.concat(data_frames, ignore_index=True)
    
    print("Done ||| Files: {} || Number of items: {}".format(count, rows))
    return df

# Example usage
filepath = '/Users/rishabkhuba/Documents/MLOps_Project/IE7374-Sepsis-Classification/notebooks/batch_experiment/batch_test_3/'  # Replace with your actual filepath
df = pd.DataFrame()  # Initial empty DataFrame
dataset = read_data(filepath, df)
```

```python
def data_preprocess_pipeline(df):

    """
    Load, preprocess, and save the dataframe.

    Args:
        Dataframe (pd.DataFrame): dataframe with features and target
    Return:
        Dataframe (pd.DataFrame): X preprocessed dataframe containing features is returned  
        Dataframe (pd.DataFrame): y series returned  
    """

    try:

        # Define columns to drop initially and after preprocessing
        initial_columns_to_drop = ['SBP', 'DBP', 'EtCO2', 'BaseExcess', 'HCO3',
                                   'pH', 'PaCO2', 'Alkalinephos', 'Calcium',
                                   'Magnesium', 'Phosphate', 'Potassium', 'PTT',
                                   'Fibrinogen', 'Unit1', 'Unit2']
        columns_with_high_nulls = ['TroponinI', 'Bilirubin_direct', 'AST', 'Bilirubin_total',
                                   'Lactate', 'SaO2', 'FiO2', 'Unit', 'Patient_ID']
        columns_to_normalize = ['MAP', 'BUN', 'Creatinine', 'Glucose', 'WBC', 'Platelets']
        
        # Create 'Unit' column from 'Unit1' and 'Unit2'
        df["Unit"] = df["Unit1"] + df["Unit2"]

        # Drop initial redundant columns
        df.drop(columns=initial_columns_to_drop, inplace=True)

        # Group by 'Patient_ID' and fill missing values
        df = df.groupby('Patient_ID').apply(lambda x: x.bfill().ffill()).reset_index(drop=True)

        # Drop columns with more than 25% null values
        df.drop(columns=columns_with_high_nulls, inplace=True)

        # Apply log transformation to normalize specific columns
        df[columns_to_normalize] = df[columns_to_normalize].apply(np.log1p)

        # One-hot encode the 'Gender' column
        df['gender'] = df['Gender'].replace({1: 'M', 0: 'F'}).apply(lambda x: x if x in ['M', 'F'] else np.nan)
        encoded_gender = pd.get_dummies(df["gender"])
        df = df.join(encoded_gender).drop(columns=["Gender", "gender"])

        # Ensure the encoded columns are integers
        df['F'] = df['F'].astype(int)
        df['M'] = df['M'].astype(int)

        # Drop remaining rows with any NaN values
        df.dropna(inplace=True)

        # Split the DataFrame back into X and y
        X_preprocessed = df.drop(columns=['SepsisLabel'])
        y_preprocessed = df['SepsisLabel']

        # Remove unwanted columns
        X_preprocessed = X_preprocessed.loc[:, ~X_preprocessed.columns.str.contains('^Unnamed', case=False, regex=True)]
        X_preprocessed.drop(columns=['Hour'], inplace=True)

        # Display shape and value counts for debugging
        print(f"Shape of X after preprocessing: {X_preprocessed.shape}")
        print(f"Shape of y after preprocessing: {y_preprocessed.shape}")
    
        return X_preprocessed, y_preprocessed
        
    except KeyError as ke:
        print("KeyError during preprocessing: %s", ke)
    except ValueError as ve:
        print("ValueError during preprocessing: %s", ve)
    except Exception as ex:
        print("An unexpected error occurred during preprocessing: %s", ex)
        raise
```

```python
def scale_test_data(X, scalar_path):
    """
    Load, scale, and save the test data using the pre-fitted scaler.

    The function:
    - Loads the processed test data from a pickle file.
    - Loads the scaler object from a pickle file.
    - Scales specified columns using the pre-fitted StandardScaler.
    - Saves the scaled test data to a pickle file.
    """
    try:
        scaler = load_data_from_pickle(scalar_path)
        columns_to_scale = ['HR', 'O2Sat', 'Temp', 'MAP', 'Resp', 'BUN', 'Chloride', 'Creatinine', 'Glucose', 'Hct', 'Hgb', 'WBC', 'Platelets']
        X[columns_to_scale] = scaler.transform(X[columns_to_scale])
        
        return X
        
    except FileNotFoundError as e:
        print(f"File not found: {e.filename}. Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred during scaling the test data: {e}")
```

```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

def predict(X_scaled, y, model_path):

    # print(model_path)
    # print("X_scaled head:")
    # print(X_scaled.head())
    
    # Load the model
    model = load_data_from_pickle(model_path)
    
    # Predict
    predictions = model.predict(X_scaled)
    
    # Calculate and print metrics
    accuracy = accuracy_score(y, predictions)
    precision = precision_score(y, predictions, average='weighted')
    recall = recall_score(y, predictions, average='weighted')
    f1 = f1_score(y, predictions, average='weighted')
    conf_matrix = confusion_matrix(y, predictions)
    class_report = classification_report(y, predictions)
    
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 Score: {f1}')
    print(f"\nConfusion Matrix:\n{conf_matrix}")
    print(f"\nClassification Report:\n{class_report}")
    
    return predictions
```

```python
y_pred = predict(X_scaled, y, model_path)
```
